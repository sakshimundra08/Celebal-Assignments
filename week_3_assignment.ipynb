```python
df = df.dropna()
```

```python
df['Stage_fear'] = df['Stage_fear'].astype('category')
df['Drained_after_socializing'] = df['Drained_after_socializing'].astype('category')
df['Personality'] = df['Personality'].astype('category')
```

```python
plt.scatter(df['Time_spent_Alone'], df['Friends_circle_size'], color='blue', alpha=0.6)
plt.title('Time Spent Alone vs Friends Circle Size')
plt.xlabel('Time Spent Alone')
plt.ylabel('Friends Circle Size')
plt.grid(True)
plt.show()
```

```python
df['Personality'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Personality Type Distribution')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
df.groupby('Personality')['Time_spent_Alone'].mean().plot(kind='bar', color=['blue', 'orange'])
plt.title('Average Time Spent Alone by Personality')
plt.ylabel('Time Spent Alone')
plt.grid(True)
plt.show()
```

```python
df.groupby('Personality')['Friends_circle_size'].mean().plot(kind='bar', color=['green', 'red'])
plt.title('Average Friends Circle Size by Personality')
plt.ylabel('Friends Circle Size')
plt.grid(True)
plt.show()
```

```python
introverts = df[df['Personality'] == 'Introvert']
extroverts = df[df['Personality'] == 'Extrovert']
```

```python
features = ['Time_spent_Alone', 'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']

for feature in features:
    plt.figure(figsize=(7, 5))
    plt.hist(introverts[feature], bins=15, alpha=0.6, label='Introvert', color='blue', edgecolor='black')
    plt.hist(extroverts[feature], bins=15, alpha=0.6, label='Extrovert', color='orange', edgecolor='black')
    plt.title(f'{feature} by Personality')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid(True)
    plt.show()
```
